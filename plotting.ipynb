```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(1,11) 
y = 2 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
#Distibution Plots
#For Numerical Variable
#distplot - univariate analysis
#joinplot - bivariate analysis
#pairplot or scatterplot    - More than 3 feautures
#Corelation helps to find out the interrelationship between all the numerical variable feautures. value range -1 to 1
#df.dtypes - to check the datatypes.
#if correlation value is positive , it means if one is increasing other also will be increasing.

import seaborn as sns
df.corr()
sns.heatmap(df.corr(),annot = True)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
#kind it means the structure will be of hexagon
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
#PDE - probability Density Functions.
```

```python
sns.pairplot(df)
#feauture need to be integer or float value.Its is also called scatter plot
```

```python
sns.pairplot(df, hue='sex')
```

```python
#Distplot help to check the distribution of the columns feautures
sns.distplot(df['tip'], kde = True, bins = 10, hist = True)
#if KDE is True y axis value will be in percentage.
```

```python
#Categorial Plots
#countplot
#boxplot
#barplot
#violinplot
# default it takes x axis
# Either x or y value other will be count

sns.countplot(df['sex'], data = df)
sns.countplot(df['smoker'],data = df)
```

```python
sns.countplot(y = df['sex'], data = df)
```

```python
#bivariate
sns.barplot(y = 'total_bill', x = 'sex', data = df)
```

```python
#boxplot
#Also worked for numerical feautures

sns.boxplot(y = 'total_bill', x = 'sex', data = df, palette ='rainbow')
```

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df, palette ='rainbow')
```
